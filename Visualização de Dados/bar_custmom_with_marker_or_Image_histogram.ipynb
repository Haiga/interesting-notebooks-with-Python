```
#@title Using Scatter Plot With image as Histogram Plot

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12,8))


total_bars = 4.5
#0.21  o espaamento das barras de bolinhas
x = np.arange(0, total_bars, 0.23)
y = np.random.uniform(0.5, 3, x.size)



all_positions = []
all_balls = []
all_size_balls = []
#0.25  o espaamento das bolinhas em uma barra
#400  o tamanho da bolinha
#0.13  o meio ba bolinha da base
def my_bar_with_one_marker(positionx, valuey, interval=0.25):
    global all_positions
    global all_balls
    global all_size_balls
    if valuey > 0:
        balls = np.arange(0.13, valuey, interval)
    else:
        balls = -1*np.arange(0.13, -1*valuey, interval)
    positions = np.array([positionx]*balls.size)
    size_balls = [400]*balls.size
    all_positions = np.concatenate((all_positions, positions))
    all_balls = np.concatenate((all_balls, balls))
    all_size_balls = np.concatenate((all_size_balls, size_balls))

    # plt.scatter(positions, balls, size_balls, c=colorRate*balls)


for xx, yy in zip(x, y):
    # plt.bar(xx, yy, color='b')
    my_bar_with_one_marker(xx, yy)


all_balls_c = np.array(all_balls)

# some_outliers = np.random.randint(0, all_balls.size, 100)
# all_balls_c[some_outliers] = -3

plt.scatter(all_positions, all_balls, all_size_balls, c='y')

#0.13 de ajuste ao grfico para a bolinha no sair do grfico
plt.xlim(np.min(x) - 0.13, np.max(x) + 0.13)
maximo = np.max(y) + 0.13
plt.ylim([0, maximo])

plt.xticks([])

#####################plot the coins
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
def getImage(path):
    return OffsetImage(plt.imread(path), zoom=0.60)

paths = ['dollar-coin.png']*all_balls.size
for x0, y0, path in zip(all_positions, all_balls, paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
    ax.add_artist(ab)
##################### put the coins above the markers

plt.show()
```

```
#@title Using Scatter Plot as Histogram Plot
import matplotlib.pyplot as plt
import numpy as np

fig = plt.subplots()


total_bars = 20

x = np.arange(total_bars)
y = np.random.uniform(0, 3, total_bars)

y_down = np.random.uniform(-3, 0, total_bars)


def my_bar_with_many_markers(positionx, valuey, interval=0.1, c='b'):
    if valuey > 0:
        balls = np.arange(0, valuey, interval)
    else:
        balls = np.arange(valuey, 0, interval)
    positions = np.array([positionx]*balls.size)
    size_balls = [10]*balls.size
    plt.scatter(positions, balls, size_balls, c=c)
    plt.scatter(positions+0.2, balls, size_balls, c=c)
    plt.scatter(positions-0.2, balls, size_balls, c=c)
    # plt.scatter(positions-0.2, balls, size_balls, c='b', marker='*')
    # plt.scatter(positions-0.2, balls, size_balls, c='b', marker='$oi$')

def my_bar_with_one_marker(positionx, valuey, interval=0.12, c='b'):
    if valuey > 0:
        balls = np.arange(0, valuey, interval)
    else:
        balls = np.arange(valuey, 0, interval)
    positions = np.array([positionx]*balls.size)
    size_balls = [10]*balls.size
    plt.scatter(positions, balls, size_balls, c=c)


for xx, yy in zip(x, y):
    # plt.bar(xx, yy, color='b')
    my_bar_with_one_marker(xx, yy, c='#000000')


for xx, yy in zip(x, y_down):
    my_bar_with_one_marker(xx, yy, c='k')
    # plt.bar(xx, yy, color='k')

plt.xlim(-1, total_bars)
minimo = np.min(y_down) - 0.1
maximo = np.max(y) + 0.1
plt.ylim([minimo, maximo])

plt.fill_between([-1, total_bars/3], [maximo, maximo], color='b', alpha=0.2)
plt.fill_between([-1, total_bars/3], [minimo, minimo], color='b', alpha=0.2)

plt.fill_between([total_bars/3, 2*total_bars/3], [maximo, maximo], color='g', alpha=0.2)
plt.fill_between([total_bars/3, 2*total_bars/3], [minimo, minimo], color='g', alpha=0.2)

plt.fill_between([2*total_bars/3, total_bars], [maximo, maximo], color='r', alpha=0.2)
plt.fill_between([2*total_bars/3, total_bars], [minimo, minimo], color='r', alpha=0.2)

plt.xticks([])
plt.show()
```
