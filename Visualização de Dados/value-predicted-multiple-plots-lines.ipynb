```
import numpy
import matplotlib.pyplot as plt

y_predict = numpy.random.uniform(low=0,high=1,size=100)
y = numpy.random.normal(loc=0.5,scale=0.25,size=100)
x = [x+1 for x in range(100)]
```

```
plt.hist(y)
```

```
plt.plot(x, y)
```

```
y.sort()
y_predict.sort()
plt.plot(x, y)
plt.plot(x, y_predict, '.')
```

```
total_points = 100

for i in range(total_points):
    plt.plot([x[i], x[i]], [y[i], y_predict[i]], 'r-')
    
plt.plot(x, y)
plt.plot(x, y_predict, 'k.')
```

```
total_points = 100
accumulated_error = [0]*total_points
for i in range(total_points):
    plt.plot([x[i], x[i]], [y[i], y_predict[i]], 'r-')
    accumulated_error[i] = accumulated_error[i - 1] + abs(y_predict[i] - y[i])
plt.plot(x, y)
plt.plot(x, y_predict, 'k.')
plt.plot(x, accumulated_error, 'g--')
#mas a visualizao  ruim 
#(se o erro acumulado passar, vamos mostrar s at o limite do grfico anterior -0.2<y<1.2)
```

```
import math
def myFunc(value):
    delta = 2
    pi = 3.14
    euler = 2.7
    mean = 5
    return (1*pow(euler, -(value-mean)/(2*(delta*delta))))/(math.sqrt(pi*(delta*delta)))

def myFuncForVec(array_of_values):
    return [myFunc(value) for value in array_of_values]
```

```
plt.plot(x, myFuncForVec(x))
```
