```
import matplotlib.pyplot as plt
import numpy as np
import math
# a = np.random.random((16, 16))

nLines = 200
nCols = 300

def my_sum(i, j):
    x = i - nLines/2
    y = j - nCols/2
    return np.sqrt(x*x + y*y)

a = np.fromfunction(my_sum, (nLines, nCols), dtype=np.float32)


plt.imshow(a, cmap='plasma', interpolation='nearest')
plt.xticks([])
plt.yticks([])
plt.show()
#using interpolation
```
